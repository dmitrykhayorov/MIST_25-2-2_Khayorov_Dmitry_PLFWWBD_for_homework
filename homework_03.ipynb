## Задача 1

```python
def count_word_frequency(text):
    text = text.lower()
    
    for char in ".,!?;:": 
        text = text.replace(char, "")
    
    words = text.split()
    
    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
            
    return frequency

# Пример использования:
sample_text = "Это просто пример. Просто очень простой пример текста для примера."
result = count_word_frequency(sample_text)
print(result)
```

## Задача 2

```python
def calculate_price(price, discount=0, tax_rate=0.2):
    price_with_discount = price * (1 - discount / 100)
    
    final_price = price_with_discount * (1 + tax_rate)
    
    return final_price

# Примеры использования:
# 1. Только цена (скидка 0%, налог 20%)
print(calculate_price(100))  
# Результат: 120.0 (100 -> 120)

# 2. Цена 100, скидка 10%, налог 20%
print(calculate_price(100, discount=10))  
# Результат: 108.0 (100 - 10 = 90; 90 + 20% = 108)

# 3. Цена 100, скидка 50%, налог 10%
print(calculate_price(100, discount=50, tax_rate=0.3))  
# Результат: 55.0

```

## Задача 3

```python
import datetime

def analyze_log(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()
    except FileNotFoundError:
        return {}

    counts = {}
    min_date = None
    max_date = None
    
    # Формат даты, ожидаемый в логах
    date_fmt = "%Y-%m-%d %H:%M:%S"

    for line in lines:
        line = line.strip()
        if not line:
            continue

        try:
            parts = line.split(',')
            
            if len(parts) < 3:
                raise IndexError
            
            date_str = parts[0].strip()
            level = parts[1].strip()
            
            current_dt = datetime.datetime.strptime(date_str, date_fmt)
            

            counts[level] = counts.get(level, 0) + 1
            
  
            if min_date is None or current_dt < min_date:
                min_date = current_dt
            
            if max_date is None or current_dt > max_date:
                max_date = current_dt
                
        except (IndexError, ValueError):
            continue
            
    # Формируем итоговый результат
    result = {
        'counts': counts,
        # Если ни одной валидной строки не найдено, диапазон останется None
        'time_range': None 
    }
    
    # Если даты были найдены, форматируем их обратно в строки
    if min_date and max_date:
        result['time_range'] = (
            min_date.strftime(date_fmt), 
            max_date.strftime(date_fmt)
        )
        
    return result

print(analyze_log("./homework.log"))
```
