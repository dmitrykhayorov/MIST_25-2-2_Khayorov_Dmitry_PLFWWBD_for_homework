## Задача 1

```python
N = int(input())

for i in range(1, N+1):
    if (i % 3 == 0) and (i % 5 == 0):
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")

    
```

## Задача 2

```python
shopping_list = []


while True:
    action = int(input("1. Добавить 2. Показать 3. Выйти\n"))
    if action == 1:
        product_name = input("Введите название продукта:\n")
        # можно обработать пустой ввод
        shopping_list.append(product_name)
    elif action == 2: 
        if (len(shopping_list) > 0):
            for i, product in enumerate(shopping_list):
                print(f"{i+1}. {product}")
        else:
            print("Список покупок пуст.")
    elif action == 3:
        print("Программа завершена.")
        break
    else:
        print("Неизвестная команда, попробуйте снова.")
        
            
        
    
```
